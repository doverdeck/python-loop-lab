```python
test_scores=[22, 99, 102, 33, 57, 75, 100, 81, 62, -56]
stored_number=0
def lowest_number(num_list):
    stored_number=num_list[0]
    for i in num_list:
        if i<stored_number:
            stored_number=i   
    return stored_number
print(lowest_number(test_scores))
print(lowest_number([48,10,4,2,7,1,9,5,6,8,3]))
print(lowest_number([432,234,234,134,13436,123,111,567]))
```
